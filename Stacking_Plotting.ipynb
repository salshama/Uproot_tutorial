```python
from convertColor import convert_color
import ROOT
```

```python
colours = ['#20b2aa','#9c3ed6','#00ff7f' ] #color-hex
root_cols = [convert_color(col,'root') for col in colours]
```

```python
hists = [ROOT.TH1D("h"+str(i),"",10,0,1) for i in range(0,3)]
for i,hist in enumerate(hists):
    hist.Sumw2()
    hist.FillRandom("gaus",10*(i+1))
    hist.SetLineColor(root_cols[i])
    hist.SetMarkerColor(root_cols[i])
    hist.SetMarkerStyle(20+i)
    hist.SetFillColor(root_cols[i])
```

```python
c = ROOT.TCanvas("c","c",800,600)
c.cd()
```

```python
hists[2].Draw()
hists[1].Draw("same")
hists[0].Draw("same")
c.Draw()
```

```python
stack = ROOT.THStack("name","title;axis1;axis2")
for hist in hists:
    stack.Add(hist)
```

```python
stack.Draw("hist")
c.Draw()
```
